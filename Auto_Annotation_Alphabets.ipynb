Generating Auto-Annotations on large test dataset

```python
model = YOLO('runs/detect/train5/weights/best.pt')

results = list(model.predict('/content/drive/MyDrive/Alphabets-Auto-Annotate/X-Y-Z/test/images'))
```

```python
output_directory = '/content/drive/MyDrive/Alphabets-Auto-Annotate/X-Y-Z/test/labels/'

for r in results:
  i = r.boxes.cls.tolist()
  bbox = r.boxes.xywhn.tolist()
  if i:
    #c = 3 + int(i[0])
    c = int(i[0])
    bbox[0].insert(0, c)
    #print(bbox[0])

    name = os.path.splitext(os.path.basename(r.path))[0]  #extracts names of files without extension
    new_file_path = os.path.join(output_directory, name + ".txt")

    with open(new_file_path, 'w') as file:
      file.write(' '.join(str(x) for x in bbox[0]))
```
